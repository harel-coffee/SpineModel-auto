```python
spkeys = spdata.keys() - ["#Sp", "#Bch", "SA", "nPSD",
                          "GPHN_SNR","DiS-EMchk"]

def set_only_1par_free(spdata, parameter, 
                       keys = spkeys, method = np.mean):
    '''
    - spdata original
    - parameter in ['Dsb', 'Dss', 
    'd_interSp', 'Vtot', 'Vh', 'Ln','A1', 'GPHN_SNR', 
    'DiS-EMchk', 'Lneck', 'maxDneck', 
    'minDneck', 'meanDneck', 'Rneck', 
    'Lhead', 'maxDhead', 'minDhead', 'meanDhead', 'Ah', 
    'Rhead', 'meanDneck_trunc70pc', 'L', 'D', 'AhA0', 'Rmorph']
    '''
    assert parameter in keys
    
    # Avoid bivalued parameters like PSD of inhibitory synapses
    bivaluedpars = ["A2"]
    notupdate = [parameter]
    if parameter not in bivaluedpars:
        notupdate.extend([par for par in bivaluedpars])
        
    # Update all keys except the parameter and bivalued parameters
    new_spset = copy.deepcopy(spdata)
    for key in set(keys)-set([parameter]):
        df = new_spset[key]
        df.iloc[:] = np.ones(df.shape[0])*method(df)
    
    # Update bivalued parameters except parameter if any left
    for key in set(bivaluedpars)-set([parameter]):
        df = new_spset[key]
        sel = df.values>0.0
        df.iloc[sel] = np.ones(sel.sum())*method(df.iloc[sel])
        
    return new_spset

def set_varying_1par(spdata, parameter, par_range = None):
    new_spset = set_only_1par_free(spdata, parameter)
    
    df = new_spset[parameter]
    if par_range is None:
        ymax = df.max()
        ymin = df.min()
    else:
        assert type(par_range) is tuple
        ymax = par_range[1]
        ymin = par_range[0]
        
    n_spines = df.shape[0]
    
    df.iloc[:] = np.linspace(ymin, ymax, n_spines)

    return new_spset
```
